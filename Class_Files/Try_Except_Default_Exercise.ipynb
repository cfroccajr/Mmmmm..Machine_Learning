```python
def new_prod(a,b):
    return int(a+b-a*b)
```

```python
def new_sum2(a=1,b=1):
    s = 1
    try:
        if type(a) != int or type(b) != int:
            raise TypeError
        else:
            s = int(a+b)
    except TypeError:
        print("Your inputs are the wrong data type, a=%s is %s and b=%s is %s." 
              % (str(a), type(a), str(b), type(b)))
        s = TypeError
    except Exception as Err:
        print("There was an error, check your inputs.\n", Err)
        s = Err
    return s
```

```python
A_inputs = [2,1,0,"t",[2,3],3,7]
B_inputs = [0,1,"a",5,[2,3],{},6]
```

```python
for a in A_inputs:
    print("\t BEGIN ROUND")
    for b in B_inputs:
        print("With inputs a=%s and b=%s we get %s" % (a,b,new_sum2(a,b)))
    print("\t NEXT ROUND")
```

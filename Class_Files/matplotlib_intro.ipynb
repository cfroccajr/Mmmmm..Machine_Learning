```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Plots with Matplotlib

```python
X = np.linspace(-5,5,100)
figure1, axis1 = plt.subplots()
axis1.plot(X,X);
```

```python
axis1.plot(X,X**2);
figure1
```

```python
axis1.set_title("My First Plot")
axis1.legend(['$x$', '$x^2$'])
axis1.set_ylim(-10,10)
figure1
```

```python
figure2, axis2 = plt.subplots(3,5,figsize=(18,10))
for row in range(3):
    for col in range(5):
        power = 5*row+col
        axis2[row,col].plot(X,X**power)
        axis2[row,col].set_ylim(-10,10)
        axis2[row,col].legend(['$x^{%s}$'%power])
```

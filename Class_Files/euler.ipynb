```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def df(x):
    return 3*x**2-4*x+1
```

```python
def f(x):
    return x**3-2*x**2+x
```

```python
x = [0.5]
y = [f(0.5)]
l = 0.5
i=0
while i<20:
    new_x = x[i]-l*df(x[i])
    new_y = f(new_x)
    if new_y < y[i]:
        x.append(new_x)
        y.append(new_y)
        l = l*1.25
        i+=1
    else:
        l = 0.75*l
print(x)
print(y)
```

```python
X = np.array(x)
Y = np.array(y)

vals = np.linspace(0,1.1,100)

plt.plot(X,Y,"ro")
plt.plot(vals,f(vals))
```

```python
# ================================================================
#   MAT 531 – Multivariable Mathematics for Machine Learning
#   Chapter 5 – Project A
#
#   Title: Power Method for Finding Eigenvalues
#   Student: Nikita S. Karim
#   Instructor Professor Rocca
#   Fall 2025
# ================================================================

import numpy as np
# Matrix A 
A = np.array([
    [6, -4, -8],
    [-3, 10, 11],
    [2, -4, -4]
], dtype=float)

x = np.array([1.0, 0.0, 0.0])

# Power Method iterations
for i in range(20):
    y = A @ x                   # multiply by A
    x = y / np.linalg.norm(y)   # normalize
    lam = x.T @ A @ x          
    print(f"Iter {i}: λ ≈ {lam}, x = {x}")
```
